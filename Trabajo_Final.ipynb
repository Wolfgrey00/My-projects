```python
class Producto:
    def __init__(self, referencia, nombre, precio_unitario, descripción):
        self.referencia = referencia
        self.nombre = nombre
        self.precio_unitario = precio_unitario
        self.descripción = descripción

    def __str__(self):
        return f"Referencia: {self.referencia}\nNombre: {self.nombre}\nPrecio Unitario: {self.precio_unitario}\nDescripción: {self.descripción}"

class Libro(Producto):
    def __init__(self, referencia, nombre, precio_unitario, descripción, ISBN, autor, existencias, activo):
        super().__init__(referencia, nombre, precio_unitario, descripción)
        self.ISBN = ISBN
        self.autor = autor
        self.existencias = existencias
        self.activo = activo

    def __str__(self):
        return f"{super().__str__()}\nISBN: {self.ISBN}\nAutor: {self.autor}\nExistencias: {self.existencias}\nActivo: {self.activo}"

    def incrementar(self, cantidad):
      if self.activo == "Activo":
          self.existencias += cantidad

    def decrementar(self, cantidad):
      if self.activo == "Activo":
        if self.existencias >= cantidad:
            self.existencias -= cantidad
        else:
            print("No hay suficientes existencias para decrementar.")

    def mostrar(self):
      return f"Existencias: {self.existencias}"

    def activar(self):
      self.activo = "Activo"

class Alimento(Producto):
      def __init__(self, referencia, nombre, precio_unitario, descripción, productor, distribuidor, existencias, activo):
        super().__init__(referencia, nombre, precio_unitario, descripción)
        self.productor = productor
        self.distribuidor = distribuidor
        self.existencias = existencias
        self.activo = activo

      def __str__(self):
        return f"{super().__str__()}\nProductor: {self.productor}\nDistribuidor: {self.distribuidor}\nExistencias: {self.existencias}\nActivo: {self.activo}"

      def incrementar(self, cantidad):
        if self.activo == "Activo":
            self.existencias += cantidad

      def decrementar(self, cantidad):
        if self.activo == "Activo":
            if self.existencias >= cantidad:
                self.existencias -= cantidad
            else:
                print("No hay suficientes existencias para decrementar.")

      def mostrar(self):
        return f"Existencias: {self.existencias}"

      def activar(self):
        self.activo = "Activo"

class Computo(Producto):
    def __init__(self, referencia, nombre, precio_unitario, descripción, marca, tipo, existencias, activo):
        super().__init__(referencia, nombre, precio_unitario, descripción)
        self.marca = marca
        self.tipo = tipo
        self.existencias = existencias
        self.activo = activo

    def __str__(self):
        return f"{super().__str__()}\nMarca: {self.marca}\nTipo: {self.tipo}\nExistencias: {self.existencias}\nActivo: {self.activo}"

    def incrementar(self, cantidad):
        if self.activo == "Activo":
            self.existencias += cantidad

    def decrementar(self, cantidad):
        if self.activo == "Activo":
            if self.existencias >= cantidad:
                self.existencias -= cantidad
            else:
                print("No hay suficientes existencias para decrementar.")

    def mostrar(self):
        return f"Existencias: {self.existencias}"

    def activar(self):
        self.activo = "Activo"

def Alta():
    print("¿Qué tipo de producto va a ser?")
    print("1. Libro")
    print("2. Alimento")
    print("3. Computo")
    opcion = int(input())
    if opcion == 1:
        Alta_Libro()
    elif opcion == 2:
        Alta_Alimento()
    elif opcion == 3:
        Alta_Computo()

def Alta_Libro():
    print("Ingrese los datos del producto")
    referencia = input("Ingrese la referencia: ")
    if referencia in listaObjetos:
        print("La referencia ya existe")
        return
    nombre = input("Ingrese el nombre: ")
    precio_unitario = float(input("Ingrese el precio unitario: "))
    descripción = input("Ingrese la descripción: ")
    ISBN = input("Ingrese el ISBN: ")
    autor = input("Ingrese el autor: ")
    existencias = int(input("Ingrese las existencias: "))
    activo = input("Ingrese el estado del producto: ")
    listaObjetos.append(Libro(referencia, nombre, precio_unitario, descripción, ISBN, autor, existencias, activo))
    print("Producto dado de alta")
    print('********************')

def Alta_Alimento():
    print("Ingrese los datos del producto")
    referencia = input("Ingrese la referencia: ")
    if referencia in listaObjetos:
        print("La referencia ya existe")
        return
    nombre = input("Ingrese el nombre: ")
    precio_unitario = float(input("Ingrese el precio unitario: "))
    descripción = input("Ingrese la descripción: ")
    productor = input("Ingrese el productor: ")
    distribuidor = input("Ingrese el distribuidor: ")
    existencias = int(input("Ingrese las existencias: "))
    activo = input("Ingrese el estado del producto: ")
    listaObjetos.append(Alimento(referencia, nombre, precio_unitario, descripción, productor, distribuidor, existencias, activo))
    print("Producto dado de alta")
    print('********************')

def Alta_Computo():
    print("Ingrese los datos del producto")
    referencia = input("Ingrese la referencia: ")
    if referencia in listaObjetos:
        print("La referencia ya existe")
        return
    nombre = input("Ingrese el nombre: ")
    precio_unitario = float(input("Ingrese el precio unitario: "))
    descripción = input("Ingrese la descripción: ")
    marca = input("Ingrese la marca: ")
    tipo = input("Ingrese el tipo: ")
    existencias = int(input("Ingrese las existencias: "))
    activo = input("Ingrese el estado del producto: ")
    listaObjetos.append(Computo(referencia, nombre, precio_unitario, descripción, marca, tipo, existencias, activo))
    print("Producto dado de alta")
    print('********************')

def Activar_Producto():
    print("Ingrese la referencia del producto a activar")
    referencia = input()
    for i in listaObjetos:
        if i.referencia == referencia:
            i.activar()
            print("Producto activado")

def Altas_de_existencias():
    print("Ingrese la referencia del producto a incrementar")
    referencia = input()
    for i in listaObjetos:
        if i.referencia == referencia:
            cantidad = int(input("Ingrese la cantidad a incrementar"))
            i.incrementar(cantidad)

def Bajas_de_existencias():
    print("Ingrese la referencia del producto a decrementar")
    referencia = input()
    for i in listaObjetos:
        if i.referencia == referencia:
            cantidad = int(input("Ingrese la cantidad a decrementar"))
            i.decrementar(cantidad)

def Consultar():
    for i in listaObjetos:
        print(i.nombre)
        print(i.mostrar())


def Mostrar_todo():
    for i in listaObjetos:
        print('******************')
        print("Producto: " + str(listaObjetos.index(i) + 1))
        print(i)


def Guardar_Archivo():
    f = open("Productos.txt", "wt")
    x = 0
    for i in listaObjetos:
        cadena = str(i)
        f.write(cadena)
        f.write("\n")
        x = x + 1
    f.close()
    print("El número de registros procesados son: ", x)
    print("Proceso Concluido")
    print("***************")

def Leer_Archivo():
    try:
        file = open('Productos.txt', 'r')
        mensaje = file.read().split('\n')
        for palabra in mensaje:
            print(palabra)
    except FileNotFoundError:
        print("Error de lectura del archivo")
    finally:
        file.close()
        print("Lectura finalizada")
```
