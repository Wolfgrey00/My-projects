```python
class Producto:
  def __init__(self, nombre, proveedor):
    self.nombre = nombre
    self.proveedor = proveedor

  def __str__(self):
    return f"Nombre: {self.nombre}\nProveedor: {self.proveedor}"

class Refresco(Producto):
  def __init__(self, nombre, proveedor, tipo, costo, existencias):
    super().__init__(nombre, proveedor)
    self.tipo = tipo
    self.costo = costo
    self.existencias = existencias

  def __str__(self):
    return f"{super().__str__()}\nTipo: {self.tipo}\nCosto: {self.costo}\nExistencias: {self.existencias}"

  def vender(self, cantidad):
    self.existencias -= cantidad

class Enlatado(Producto):
  def __init__(self, nombre, proveedor, tipo, costo, existencias):
    super().__init__(nombre, proveedor)
    self.tipo = tipo
    self.costo = costo
    self.existencias = existencias

  def vender(self, cantidad):
    self.existencias -= cantidad

  def __str__(self):
    return f"{super().__str__()}\nTipo: {self.tipo}\nCosto: {self.costo}\nExistencias: {self.existencias}"


#Ejemplo de uso
refresco = Refresco("Coca-Cola", "Coca-Cola Company", "Cola", 15, 100)
enlatado = Enlatado("Atún Dolores", "Dolores", "Atún en agua", 25, 50)

print("Refresco:")
print("Nombre:", refresco.nombre)
print("Proveedor:", refresco.proveedor)
print("Tipo:", refresco.tipo)
print("Costo:", refresco.costo)
print("Existencias:", refresco.existencias)

print("\nEnlatado")
print("Nombre:", enlatado.nombre)
print("Proveedor:", enlatado.proveedor)
print("Tipo:", enlatado.tipo)
print("Costo:", enlatado.costo)
print("Existencias:", enlatado.existencias)


print(refresco)
print("**************")
print(enlatado)

enlatado.vender(10)
print(enlatado)

refresco.vender(20)
print(refresco)

listaProductos=[]
listaProductos.append(refresco)
listaProductos.append(enlatado)
listaProductos.append(Refresco("Pepsi", "Pepsico Company", "Cola", 20, 150))
listaProductos.append(Enlatado("Span", "FUD", "Jamon", 40, 500))


for producto in listaProductos:
  print("**************")
  print(producto)

while True:
   nom = input("Indica el nombre del producto: ")
   prov = input("Indica el proveedor del producto:")
   tipo = input("Indica el tipo del producto:")
   costo = float(input("Indica el costo del producto:"))
   existencias = int(input("Indica las existencias del producto:"))
   tp = int(input("Indica el tipo de producto 1. Refresco, 2. Enlatados:"))
   if tp == 1:
     listaProductos.append(Refresco(nom, prov, tipo, costo, existencias))
   else:
     listaProductos.append(Enlatado(nom, prov, tipo, costo, existencias))
   print("Producto agregado")
   salir = input("Desea agregar uno nuevo: S (Sí) / N (No): ")
   if salir == "N" or salir == 'n':
     break
```

```python
f = open("Productos.txt", "a+")
x = 0
for i in listaProductos:
    cadena = str(i)
    f.write(cadena)
    f.write("\n")
    x = x + 1
f.close()
print("El número de registros procesados son: ", x)
print("Proceso Concluido")
print("***************")
print("Lectura del archivo")
x = 1
y = 0
try:
    file = open('Productos.txt', 'r')
    mensaje = file.read().split('\n')
    for palabra in mensaje:
        print(palabra)
except FileNotFoundError:
    print("Error de lectura del archivo")
finally:
    file.close()
```
